## Intuition[^cookiecutter_Other people will thank you]

A well-defined, standard project structure means that a newcomer can begin to understand an analysis without digging in to extensive documentation. It also means that they don't necessarily have to read 100% of the code before knowing where to look for very specific things.

Well organized code tends to be self-documenting in that the organization itself provides context for your code without much overhead. People will thank you for this because they can:

- Collaborate more easily with you on this analysis
- Learn from your analysis about the process and the domain
- Feel confident in the conclusions at which the analysis arrives

Not only is a well-defined project structure important for others, it is also important for you. You don't want scripts like `train.py` and `train1.py` lying in the same directory, what does the suffix $1$ mean?

Overall, having a well-defined and organized code structure is integral not only for yourself, but for other team members.

### Config and Initialization

In this section, we will be creating `config.py`, a configuration file to store global variables and also **initiate and create** folders.

```python
!mkdir -p config
CONFIG_DIR = Path.joinpath(BASE_DIR, "config")
```

```python
%%writefile {CONFIG_DIR}/config.py
import logging
import sys
from dataclasses import dataclass
from pathlib import Path
from typing import List, Optional

# Creating Directories
BASE_DIR = Path(__file__).parent.parent.absolute()

CONFIG_DIR = Path(BASE_DIR, "config")
LOGS_DIR = Path(BASE_DIR, "logs")
SRC_DIR = Path(BASE_DIR, "src")
DATA_DIR = Path(BASE_DIR, "data")
STORES_DIR = Path(BASE_DIR, "stores")
TEST_DIR = Path(BASE_DIR, "tests")

## Local stores
MODEL_REGISTRY = Path(STORES_DIR, "model")
RAW_DATA = Path(DATA_DIR, "raw")
PROCESSED_DATA = Path(DATA_DIR, "processed")

## Create dirs
for d in [
    CONFIG_DIR,
    LOGS_DIR,
    DATA_DIR,
    STORES_DIR,
    TEST_DIR,
    MODEL_REGISTRY,
    RAW_DATA,
    PROCESSED_DATA,
]:
    d.mkdir(parents=True, exist_ok=True)
```

```python
%%writefile {BASE_DIR}/main.py
from config import config
```

```python
!python3 main.py # call main to create the folders since it calls config!
```
